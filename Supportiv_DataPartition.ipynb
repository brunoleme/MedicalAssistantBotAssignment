# Loading References


```python
import pandas as pd
```

Loading CSV file with all questions and answer available.

```python
df = pd.read_csv('intern_screening_dataset.csv')
```

```python
df
```

Implementing a logic to keep the last reference of each question, as part of the evaluation dataset, the previous references will be considered as part of the training set. Except when the question has only one reference, in this situation it will be included in the training dataset.

```python
grouped = df.groupby('question')
```

```python
train_data = []
eval_data = []
```

```python
for question, group in grouped:
    if len(group) > 1:
        train_data.extend(group[:-1].to_dict('records'))
        eval_data.extend(group[-1:].to_dict('records'))
    else:
        train_data.extend(group.to_dict('records'))

train_df = pd.DataFrame(train_data, columns=df.columns)
eval_df = pd.DataFrame(eval_data, columns=df.columns)
```

Exporting training and evaluation samples

```python
train_df.to_csv('intern_screening_dataset_train.csv', index=False)
eval_df.to_csv('intern_screening_dataset_eval.csv', index=False)
```
